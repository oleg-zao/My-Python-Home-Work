Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
# Фільтруємо дані для міст Yangon та Mandalay
yangon_data = df[df['City'] == 'Yangon']
mandalay_data = df[df['City'] == 'Mandalay']

# Обчислюємо середнє значення gross income для кожного міста
yangon_mean_income = yangon_data['gross income'].mean()
mandalay_mean_income = mandalay_data['gross income'].mean()

# Виводимо результати
print("Середній дохід у Yangon:", yangon_mean_income)
print("Середній дохід у Mandalay:", mandalay_mean_income)

if yangon_mean_income > mandalay_mean_income:
    print("У Yangon середній дохід вищий.")
else:
    print("У Mandalay середній дохід вищий.")
```

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
# Фільтруємо дані для міст Yangon та Mandalay
yangon_data = df[df['City'] == 'Yangon']
mandalay_data = df[df['City'] == 'Mandalay']

# Рахуємо відсотковий розподіл кількості інвойсів по категоріям товарів для міста Yangon
yangon_product_distribution = yangon_data['Product line'].value_counts(normalize=True) * 100

# Рахуємо відсотковий розподіл кількості інвойсів по категоріям товарів для міста Mandalay
mandalay_product_distribution = mandalay_data['Product line'].value_counts(normalize=True) * 100

# Виводимо результати
print("Yangon Product Distribution:")
print(yangon_product_distribution)

print("\nMandalay Product Distribution:")
print(mandalay_product_distribution)
```
