3.1. Створіть нову колонку в даних `Rating Category`, яка визначає категорію рейтингу: високий, середній чи низький за наступним правилом:
- якщо рейтинг вищий або рівний 8 - то він високий (`High`)
- якщо рейтинг від 6 до 8 не включаючи - то він середній (`Medium`)
- інакше - низький (`Low`).

```python
# Створюємо функцію для категоризації рейтингу
def categorize_rating(rating):
    if rating >= 8:
        return 'High'
    elif 6 <= rating < 8:
        return 'Medium'
    else:
        return 'Low'

# Створюємо нову колонку 'Rating Category' за допомогою функції
df['Rating Category'] = df['Rating'].apply(categorize_rating)

# Переглядаємо перші кілька рядків з новою колонкою
display(df.head())
```

3.2. Виведіть, який відсоток даних становить кожне зі значень категорій рейтингу.

```python
# Обчислюємо відсоток кожної категорії рейтингу
rating_category_percent = df['Rating Category'].value_counts(normalize=True) * 100

# Виводимо результат
print(rating_category_percent)
```

4.2. З допомогою функції `get_hour` і метода `DataFrame.apply()` створіть нову колонку `Hour` в нашому наборі даних, яка буде містити години здійснення покупки, обчислені на основі часу здійснення покупки `Time`.

```python
# Спочатку переконаємося, що функція get_hour вже визначена
def get_hour(time_str):
    # Розділяємо рядок за двокрапкою і беремо перший елемент (годину)
    hour = time_str.split(':')[0]
    # Конвертуємо його у тип int і повертаємо
    return int(hour)

# Використовуємо apply() для створення нової колонки 'Hour'
df['Hour'] = df['Time'].apply(get_hour)

# Переглянемо перші кілька рядків, щоб перевірити нову колонку
display(df.head())
```

4.3. Виведіть кількість значень за кожною годиною покупки і дайте відповідь "О яких годинах найбільше здійснюють покупок покупці і чому на Ваш погляд саме о цих годинах? Чи це слогічно, що найбільше покупок саме в ці години?".

```python
# Рахуємо кількість покупок за кожною годиною
hourly_counts = df['Hour'].value_counts().sort_index()

# Виводимо результат
display(hourly_counts)
```

З огляду на результати:

- Найбільше покупок було здійснено о **19:00** (113 покупок).
- Інші пікові години: **13:00** (103 покупок), **15:00** (102 покупок), **10:00** (101 покупка).
- Найменше покупок о **17:00** (74 покупки).

### Чому саме ці години мають найбільше покупок:
1. **19:00** — це вечірній час, коли багато людей завершують роботу, повертаються додому або мають час зайти до магазину. Це логічно, адже вечір — зручний час для покупок після робочого дня.
2. **13:00** та **15:00** — це періоди обідньої перерви та післяобіднього часу. Можливо, в ці години люди також виходять на обід або роблять короткі перерви для покупок.
3. **10:00** — це ранковий час, коли багато людей можуть робити покупки після відкриття магазинів або початку робочого дня.

### Логічність піків:
Так, це логічно, що покупки зосереджені в обідній час і вечірні години, оскільки це найзручніші періоди для багатьох покупців. Більшість людей не мають можливості робити покупки під час активної роботи або на початку дня.
