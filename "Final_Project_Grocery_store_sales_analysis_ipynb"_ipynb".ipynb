### **3. Які категорії найбільш прибуткові? (Парето-аналіз)**

```python
# Розподіл товарів по категоріям
profit_by_category = df.groupby('Category')['Profit Margin'].sum().reset_index()
profit_by_category = profit_by_category.sort_values(by='Profit Margin', ascending=False)
```

```python
# Кумулятивний процент
total_profit = profit_by_category['Profit Margin'].sum()
profit_by_category['Cumulative Profit'] = profit_by_category['Profit Margin'].cumsum()
profit_by_category['Cumulative Percentage'] = 100 * profit_by_category['Cumulative Profit'] / total_profit
```

```python
# Будуємо график
fig, ax1 = plt.subplots(figsize=(15, 5))

# Колонки для прибутку
color = 'tab:blue'
ax1.set_xlabel('Category')
ax1.set_ylabel('Profit Margin', color=color)
ax1.bar(profit_by_category['Category'], profit_by_category['Profit Margin'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Лінія для кумулятивного проценту
ax2 = ax1.twinx()  # Друга вісь для лінії
color = 'tab:red'
ax2.set_ylabel('Cumulative Percentage (%)', color=color)
ax2.plot(profit_by_category['Category'], profit_by_category['Cumulative Percentage'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# Заголовок та графік
plt.title('Парето-аналіз: Категоріі товарів за прибутком')
plt.show()
```

Графік показує залежність прибутку від категорії куплених товарів

**11. Чи залеже прибуток від знижок?**

```python
plt.figure(figsize=(8, 5))

# Побудова графіка з лінією регресії
sns.regplot(x=df['DISC'], y=df['Profit Margin'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Залежність прибутка від знижок')
plt.xlabel('Знижки')
plt.ylabel('Прибуток')

plt.show()
```

З представленого графіка можна зробити кілька спостережень:

### 1. Відсутність чіткої залежності:
   - На графіку не спостерігається чіткої лінійної або будь-якої іншої закономірної залежності між розміром знижки та прибутком. Точки розташовані вертикально для кожної величини знижки, що вказує на те, що прибуток може варіюватися незалежно від розміру знижки.

### 2. Концентрація значень прибутку біля нуля:
   - Більшість точок знаходяться близько до осі X (навколо нуля на осі Y), що може означати, що багато товарів, незалежно від розміру знижки, приносять низький прибуток або взагалі не приносять прибутку.

### 3. Викиди:
   - Можна помітити кілька викидів — точки, які значно вищі за інші. Ці викиди показують, що для деяких товарів знижки, можливо, сприяли високому прибутку, але таких випадків небагато.

### 4. Варіативність прибутку при одному рівні знижок:
   - Навіть при однаковому розмірі знижки, прибуток варіюється в широких межах. Це може вказувати на те, що інші фактори (крім знижок) відіграють значну роль у визначенні прибутку.

### Висновок:
Графік показує, що знижки самі по собі не є ключовим фактором, який визначає прибуток, або їх вплив може значно варіюватися залежно від інших факторів. Для точнішого розуміння варто провести подальший аналіз, включаючи додаткові змінні.

**12. Чи залеже прибуток від операційних витрат?**

```python
import statsmodels.api as sm

# Будуємо регресійну модель
X = df['% Operating Cost']
y = df['Profit Margin']

# Константа для моделі (змыщення)
X = sm.add_constant(X)

# Будуємо модель
model = sm.OLS(y, X).fit()
print(model.summary())

# Візуалізація результатів
plt.figure(figsize=(7, 5))
sns.regplot(x='% Operating Cost', y='Profit Margin', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Залежність прибутку від операційних витрат')
plt.xlabel('Операційні витрати (%)')
plt.ylabel('Прибуток (%)')
plt.show()
```

**Висновок**:

1.   Слабка залежність: оскільки операційні витрати варіюються в дуже вузькому діапазоні (вони постійні), важко виявити значну залежність між операційними витратами та прибутком.
2.   Відсутність кореляції: можливо, операційні витрати насправді не мають значного впливу на прибуток, особливо при такому вузькому діапазоні змін. Видно точки з більш високим прибутком (10-25%), що може бути викидами або особливими випадками.
