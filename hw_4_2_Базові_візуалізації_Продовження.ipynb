### Завдання: Візуалізація функцій експоненти, логарифму, і модифікованої синусоїди

**Мета:** Створити фігуру з трьома панелями: на першій показати графік експоненціальної функції, на другій - логарифмічну функцію, а на третій - модифіковану синусоїду.

1. Імпортуйте необхідні бібліотеки:


```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```


2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
# Створюємо масив x з 400 точок у діапазоні від -2 до 2
x = np.linspace(-2, 2, 400)

# Створюємо масив x_positive з 400 точок у діапазоні від 0.1 до 2
x_positive = np.linspace(0.1, 2, 400)
```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
# Будуємо фігуру Matplotlib
plt.figure(figsize=(8, 8))  # Експериментально визначений розмір

# Візуалізуємо експоненціальну функцію y = e^x
plt.subplot(3, 1, 1)  # Перша панель
plt.plot(x, np.exp(x))

# Візуалізуємо логарифмічну функцію y = ln(x)
plt.subplot(3, 1, 2)  # Друга панель
plt.plot(x, np.log(x))

# Візуалізуємо модифіковану синусоїду y = sin(x) + x
plt.subplot(3, 1, 3)  # Третя панель
plt.plot(x, np.sin(x) + x)

# Відображення графіка
plt.show()
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
plt.figure(figsize=(8, 8))

plt.subplot(3, 1, 1)  # Перша панель
plt.plot(x, np.exp(x), color='blue')  # Експоненціальна функція
plt.title('Експоненціальна функція')  # Заголовок
plt.xlabel('x')  # Мітка осі x
plt.ylabel('y')  # Мітка осі y
plt.grid(True)  # Сітка

plt.subplot(3, 1, 2)  # Друга панель
plt.plot(x, np.log(x), color='red')  # Логарифмічна функція
plt.title('Логарифмічна функція')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.subplot(3, 1, 3)  # Третя панель
plt.plot(x, np.sin(x) + x, color='green')  # Модифікована синусоїда
plt.title('Модифікована синусоїда')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Відображення графіка
plt.tight_layout()  # Автоматично налаштовує розмір підзаголовків
plt.show()
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
# Будуємо фігуру Matplotlib
plt.figure(figsize=(10, 6))

# Візуалізуємо три функції на одному графіку
plt.plot(x, np.exp(x), color='blue', label='Експоненціальна функція')
plt.plot(x, np.log(x), color='red', label='Логарифмічна функція')
plt.plot(x, np.sin(x) + x, color='green', label='Модифікована синусоїда')

# Додаємо легенду
plt.legend()

# Заголовок та мітки осей
plt.title('Графіки функцій')
plt.xlabel('x')
plt.ylabel('y')

# Відображення графіка
plt.grid(True)
plt.savefig('my_figure.png')
plt.show()
```
